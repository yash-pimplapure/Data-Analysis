```python
most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
# changing index to release date
```

```python
df_tracks.set_index('release_date',inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
# to see artist in specific index
```

```python
df_tracks[['artists']].iloc[20]
```

```python
# changing duration in seconds
```

```python
df_tracks["durations"] = df_tracks["duration_ms"].apply(lambda x : round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
df_tracks.durations.head()
```

```python
df_tracks.head()
```

```python
# correlation heatmap between variables 
```

```python
corr_df = df_tracks.drop(["key","mode","explicit"], axis = 1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df,annot = True,fmt=".1g",vmin = -1,vmax = 1,center=0,cmap="inferno",linewidths=1,linecolor="black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df , y = "loudness", x = "energy", color = "r").set(title="Loudness vs Energy Correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df , y = "popularity", x = "acousticness", color = "g").set(title="Popularity vs Acousticness Correlation")
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="NO. Of Songs Per Year")
```

```python
total_dr = df_tracks.durations
fig_dims = (18,7)
fig,ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=years, y = total_dr,ax = ax,errwidth = False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```
